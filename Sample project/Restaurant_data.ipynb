```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python

# Data structure: [restaurant_id, 2021, 2022, 2023, 2024]
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])
```

```python
print('Shape of Data is ',sales_data.shape)
print('Sample data for firsst 3 restraunts is:',sales_data[0:3])
```

## yearly total sales


```python
yearly_sales=np.sum(sales_data[:,1:],axis=0)
print(yearly_sales)
```

## Minimum sales per restauraunt


```python
min_sales=np.min(sales_data[:,1:],axis=1)
print(min_sales)
```

## Maximum sales per year

```python
max_sales=np.max(sales_data[:,1:],axis=0)
print(max_sales)
```

## Average Sales per Restauraunt


```python
avg_sales=np.mean(sales_data[:,1:],axis=1)
print(avg_sales)
```

## Cumulative sales Across Restaurants 

```python
Cumulative_sales=np.cumsum(sales_data[:,1:],axis=1)
print(Cumulative_sales)

#Plot
plt.figure(figsize=(9,6))
plt.plot(np.mean(Cumulative_sales,axis=0))
plt.title("average Cumulative sales across all restaurants")
plt.xlabel("Years")
plt.ylabel("Sales")
plt.grid(True)
plt.show()
         
```
